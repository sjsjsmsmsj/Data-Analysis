#### 1.Scatter pilot

```python
import seaborn as sns
%matplotlib inline
```

```python
tips = sns.load_dataset("tips")
```

```python
sns.scatterplot(
    data = tips, x = 'total_bill', y = 'tip'
)
```

```python
import matplotlib.pyplot as plt
def regplot(data, x, y, color='g', fontsize=14, figsize=(12, 8), title=None):
    sns.set_theme(color_codes=True)
    plt.figure(figsize = figsize)
    ax = sns.regplot(x = x, y = y, data=tips, color=color)
    plt.title(title, fontsize=fontsize)
```

```python
regplot(x ='total_bill', y = 'tip', data=tips, color='g', title='total_bill vs tip')
```

## Correlogramm

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Define the heatmap function
def heatmap(data, figsize=(12, 10), title=None, fontsize=14):
    plt.figure(figsize=figsize)
    corr = data.select_dtypes(include=['float64', 'int64']).corr()  # Select only numeric columns
    sns.heatmap(corr, xticklabels=corr.columns, yticklabels=corr.columns, cmap='RdYlGn', center=0, annot=True)
    plt.title(title, fontsize=fontsize)
    plt.show()

# Load the dataset
df = pd.read_csv("https://github.com/selva86/datasets/raw/master/mtcars.csv")

# Use the heatmap function
heatmap(df)
```

```python
-1 = 1
0: no corr
-> -1: negative
-> +1: positive
```

# 5.Pairwise Plot

```python
def heatmap(data, hue=None, figsize=(12, 10), title=None, fontsize = 14):
    plt.figure(figsize=figsize)
    sns.pairplot(data, kind='scatter', hue=hue, plot_kws=dict(s=80, edgecolor='white', linewidth=2.5))
    plt.title(title, fontsize=fontsize)
```

```python
df = sns.load_dataset('iris')
heatmap(data=df, hue='species')
```
