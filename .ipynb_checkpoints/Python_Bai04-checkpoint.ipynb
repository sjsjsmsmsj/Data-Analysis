# Arithmetic Operations

```python
import pandas as pd
import numpy as np
```

```python
s1 = pd.Series(np.arange(4),
               index=["a", "c", "d", "e"])
s2 = pd.Series(np.arange(5),
               index=["a", "c", "e", "f", "g"])
print(s1)
print(s2)
```

```python
s1 + s2
```

```python
df1 = pd.DataFrame ( np.arange(6).reshape(2, 3), columns=list("ABC"), index=["Tim", "Tom"])
df2 = pd.DataFrame(np.arange(9).reshape(3,3), columns=list("ACD"), index=["Tim", "Kate", "Tom"])
```

```python
print(df1)
print(df2)
```

```python
1 / df1
```

```python
df1 + df2
```

```python
df1.add(df2,fill_value=0)
```

```python
df1 * 3
```

```python
df1.mul(3)
```

```python
df2
```

```python
s = df2.iloc[1]
s
```

```python
df2 - s
```

## Apllying a Function

```python
df = pd.DataFrame(
    np.random.randn(4, 3),
    columns=list("ABC"),
    index=["Kim", "Susan", "Tim", "Tom"])
df
```

```python
f = lambda x: max(x) - min(x)
```

```python
df.apply(f)
```

```python
df.apply(f, axis=1)
```

## Sorting and Ranking

```python
s = pd.Series(range(5), index=["e", "d", "a", "b", "c"])
s
```

```python
s.sort_index()
```

```python
df = pd.DataFrame(
    np.arange(12).reshape(3, 4),
    index=["two", "one", "three"],
    columns=["d", "a", "b", "c"])
df
```
